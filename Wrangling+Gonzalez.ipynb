```python
import pandas as pd   # Importamos la libreria pandas 
```

```python
dfp = pd.read_csv('/content/drive/MyDrive/p_data.csv', ';', on_bad_lines='skip')   # leemos el archivo p_data.csv desde el Drive
dfw = pd.read_csv('/content/drive/MyDrive/w_data.csv', ';', on_bad_lines='skip')   # leemos el archivo w_data.csv desde el Drive 
```

```python
dfp = pd.DataFrame(dfp)   # Creamos un DataFrame con el archivo p_data.csv
dfp   # Llamamos el DataFrame
```

```python
dfw = pd.DataFrame(dfw)   #Creamos un DataFrame con el archivo w_data.csv
dfw   #Llamamos el DataFrame
```

```python
df = pd.merge(dfp, dfw, how='inner', left_on='ID', right_on='ID', indicator=False )    #Creamos un DataFrame con el archivo w_data.csv
df
```

1. Cuantos ID tienen en común. Nombrar cuales no tienen en común 

```python
dfw['ID'].isin(dfp['ID']).value_counts()    #A través de la función 'isin' se comparan las columnas 'ID' de los datasets dfw y dfp, se presentan 48842 datos que coinciden en ambas tablas y solo 1 dato que no 
```

2. Corregir los target a solo dos valores distintos '<=50k' y '>=50k' 

```python
df.replace({'Target' : '<=50K.'}, '<=50K', inplace=True)    #Se remplazan los valores '<=50K.' por '>50K'
df.replace({'Target' : '>50K.'}, '>50K', inplace=True)    #Se remplazan los valores '>50K.' por '>50'
```

3. Calcular la proporción que hay entre personas con salarios >50k respecto del totaL: Personas de distinto sexo, raza y educación. Por ejemplo, cuál es el porcentaje de mujeres con un salario mayor a 50k y compararlo con el de los hombres, lo mismo para personas de distinta raza y educación 

```python
df_pivot = pd.pivot_table(df, index='Target', columns=['Sex'], values='Capital Gain', aggfunc=sum).apply(lambda x:100*x/float(x.sum())).round(2)    #Se hace un pivot para clasificar por sexo el nivel de salarios devengados 
df_pivot
```

```python
df_pivot1 = pd.pivot_table(df, index='Target', columns=['Education'], values='Capital Gain', aggfunc=sum).apply(lambda x:100*x/float(x.sum())).round(2)   #Se realiza un pivot que relaciona el nivel de salarios con los diferentes niveles de educación
df_pivot1
```
