2. Realizar un histograma con los salarios, ¿Que rango de salarios son más populares?

```python
from matplotlib import pyplot as plt    # importamos la herramienta pyplot de la libreria matplotlib
plt.hist(df['Salary'], 15, color='blue', ec='black')    # graficamos el histograma de salarios 
```

Los rangos de salarios más populares son los que estan entre los valores de 45046-58709 y los valores de 58709-72373 

3. Realizar un gráfico de violin con los salarios discriminados por sexo

```python
import seaborn as sns   # importamos la libreria seaborn a través de la cual podremos realizar la gráfica de violin 
plt.figure()    # se llama la funcion para hacer la grafica
ax = sns.catplot(data=df, kind='violin', x='GenderID', y='Salary', hue='GenderID')    # definimos los parametros que queremos para nuestra gráfica de violin
ax.set(xlabel='Genero', ylabel='Rango de salarios', title='Salarios discriminados por genero')    # definimos los nombres de nuestra gráfica y de los ejes x, y
```

4. Graficas la serie de tiempo correspondiente a la fecha de contratación (DateofHire)

```python
Dateofhire = df.DateofHire    # separamos la columna del dataframe que contiene la información "DateofHire"
Dateofhire    # visualizamos los datos de la serie de tiempo
```

```python
df = df.set_index('DateofHire').sort_index().reset_index()    # ordenamos la serie de tiempo 
y = df.EmpID    # definimos la variable y
x = df.DateofHire   # definimos la variable x
```

```python
import matplotlib.pyplot as plt   # importamos la herramienta para graficar 
fig, ax = plt.subplots(figsize=(12, 4))  # definimos el subplot
ax.plot(x, y, label='Date of hire vs EmpID')    # deinmos el nombre de la gráfica 
ax.set_xlabel('Date of Hire')  						# Etiqueta eje x
ax.set_ylabel('EmpID')  	# Etiqueta eje y
ax.set_title('Serie de tiempo para fecha de contratación') 	# Título
ax.legend() 
```
